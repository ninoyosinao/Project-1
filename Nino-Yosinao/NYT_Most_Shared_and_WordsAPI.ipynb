```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pytrends.request import TrendReq
from pprint import pprint
import string

# Import API key
from config import api_key
```

## Generate Top Query List

```python
type_list = ["","","","","","","","","","","","","","","",""]

# Initialize empty list to hold words
words = []

# Loop through each type
for type in type_list:
    # Starting URL for NYT Most Shared API Call
    url = f"http://api.nytimes.com/svc/mostpopular/v2/mostshared/{type}/30.json?"
    query_url = f"{url}api-key={api_key}"
    response = requests.get(query_url).json()
    pprint(response)

# Loop through the list of articles, remove common punctuation, and perform a split
for result in range(len(response['results'])):
    text = response['results'][result]['title']
    text = text.replace(":","!")
    text = text.replace("!",".")
    text = text.replace(".","")
    text = text.replace("''","")
    text = text.replace("?","")
    text = text.replace("‘","")
    text = text.replace(",","")
    text = text.replace("’ll","")
    text = text.replace("’s","")
    text = text.lower()
    text_split= text.split(" ")
    for word in text_split:
        # Call the Words API
        #url = f"https://wordsapiv1.p.mashape.com/words/{word}/definitions"
        #word_json = requests.get(url).json()
        #print(word_json)
        #pos = word_json[0]["definitions"]["partOfSpeech"]
        word = word.translate(str.maketrans({":":None,"!":None,".":None, "'": None, "?": None, "‘": None, ",": None}))
        words.append(word)
print(words)
```

```python
word_df=pd.DataFrame(words)
word_df.head()
word_df_rnm = word_df.rename(columns={0:"Words"})
word_df_rnm['Words'].value_counts()
```
