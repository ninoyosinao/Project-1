```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
em_data = {'Query': ['Roman Catholic Church', 'United States Government',
                     'Lawsuit','Republican Party', 'Intelligence Service', 
                     'Election of House of Representatives',
                     'New York Times', 'US International Relation',
                     'Democratic Party', 'ISIS'], 
           'num_days': [273,253,365,154,337,28,266,238,288,14]}
em_df = pd.DataFrame(em_data)
em_df
```

```python
plt.bar(em_df['Query'],em_df['num_days'],color='g', alpha=0.5, align="center")
plt.xticks(rotation=90)
plt.show()
plt.title("Top Queries from Email on Google trend")
plt.xlabel("Query")
plt.ylabel("Number of Days over 50%")
plt.savefig("EM_Query.png", bbox_inches='tight')
```
