```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fb_data = {'Query': ['Justice Department','ISIS',
                     'Frauds and Swindling', 'Sex Crimes',
                     'Immigration',
                     'Espionage and intelligence service',
                     'Presidential election of 2016',
                     'Roman Catholic Church',
                     'Education department (us)', 
                     'Suits and Litigation (Civil)','Democratic Party'], 
           'num_days': [365, 14, 351, 365, 14,337,91,274,351,225,288]}
fb_df = pd.DataFrame(fb_data)
fb_df
```

```python
plt.bar(fb_df['Query'],fb_df['num_days'],color='b', alpha=0.5, align="center")
plt.xticks(rotation=90)
plt.show()
plt.title("Top Queries from Facebook on Google trend")
plt.xlabel("Query")
plt.ylabel("Number of Days over 50%")
plt.savefig("FB_Query.png", bbox_inches='tight')
```
