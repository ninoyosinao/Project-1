```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tw_data = {'Query': ['Intelligence Service', 'United States Government',
                     'Republican party', 'Immigration', 'Democratic Party', 
                     'Sex crimes', 'CIA', 'Missing persons',
                     'Books and Literatures', 'Justice Department', 'Roman Catholic Church',
                     'ISIS', 'Wikileaks'], 
           'num_days': [337,253,154,14,288,365,49,21,281,365,274,14,7]}
tw_df = pd.DataFrame(tw_data)
tw_df
```

```python
plt.bar(tw_df['Query'],tw_df['num_days'],color='lightblue', align="center")
plt.xticks(rotation=90)
plt.show()
plt.title("Top Queries from Twitter on Google trend")
plt.xlabel("Query")
plt.ylabel("Number of Days over 50%")
plt.savefig("TW_Query.png", bbox_inches='tight')
```
